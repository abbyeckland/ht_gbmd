```python
# Import numpy and matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Import the pymt package 
import pymt.models
# Create new instance. With each new run, it is wise to rename the instance.
hydrotrend = pymt.models.Hydrotrend()
```

```python
# Add directory for output files
config_file, config_folder = "hydro_config.txt", "Hydrotrend"
```

```python
cat  Hydrotrend/HYDRO.IN
```

```python
cat  Hydrotrend/HYDRO0.HYPS
```

```python
cat  Hydrotrend/HYDRO.CLIMATE
```

```python
#In pymt one can always find out what output a model generates by using the .output_var_names method. 
hydrotrend.output_var_names
```

```python
# Now we initialize the model with the configure file and in the configure folder
hydrotrend.initialize(config_file, config_folder)
```

```python
# this line of code lists time parameters: when, how long, and at what timestep the model simulation will work
hydrotrend.start_time, hydrotrend.time, hydrotrend.end_time, hydrotrend.time_step, hydrotrend.time_units
```

```python
# this code declares numpy arrays for several important parameters we want to save.
# empty is declaring space in memory for info to go in
n_days = int(hydrotrend.end_time)
q = np.empty(n_days) # river discharge at the outlet
qs = np.empty(n_days)# sediment load at the outlet
cs = np.empty(n_days) # suspended sediment concentration for different grainsize classes at the outlet
qb = np.empty(n_days) # bedload at the outlet
# here we have coded up the time loop using i as the index
# we update the model with one timestep at the time, untill we reach the end time 
# for each time step we also get the values for the output parameters we wish to 
for i in range(n_days):
    hydrotrend.update()
    q[i] = hydrotrend.get_value("channel_exit_water__volume_flow_rate")
    qs[i] = hydrotrend.get_value("channel_exit_water_sediment~suspended__mass_flow_rate")
    cs[i] = hydrotrend.get_value("channel_exit_water_sediment~suspended__mass_concentration")
    qb[i] = hydrotrend.get_value("channel_exit_water_sediment~bedload__mass_flow_rate")
```

```python
# Plot river discharge (q)
plt.plot(q, color='tab:blue')
plt.title('Simulated Qw')
plt.xlabel('Day in simulation')
plt.ylabel('River discharge (m3/s)')
plt.show()
```

```python
# Plot suspended sediment discharge (qs)
plt.plot(qs,color='tab:brown')
plt.title('Simulated Qs')
plt.xlabel('Day in simulation')
plt.ylabel('Sediment load (kg/s)')
plt.show()
```

```python
# Print summary statistics from model run
print('Mean')
print(q.mean(), hydrotrend.var_units("channel_exit_water__volume_flow_rate"), '= Mean Water Discharge')
print(qs.mean(), hydrotrend.var_units("channel_exit_water_sediment~suspended__mass_flow_rate"), '= Mean Suspended Sediment Discharge')
print(cs.mean(), hydrotrend.var_units("channel_exit_water_sediment~suspended__mass_concentration"), '= Mean Suspended Sediment Concentration')
print(qb.mean(), hydrotrend.var_units("channel_exit_water_sediment~bedload__mass_flow_rate"), '= Mean Bedload Discharge')
print()
print('Max')
print(q.max(), hydrotrend.var_units("channel_exit_water__volume_flow_rate"), '= Max Water Discharge')
print(qs.max(), hydrotrend.var_units("channel_exit_water_sediment~suspended__mass_flow_rate"), '= Max Suspended Sediment Discharge')
print(cs.max(), hydrotrend.var_units("channel_exit_water_sediment~suspended__mass_concentration"), '= Max Suspended Sediment Concentration')
print(qb.max(), hydrotrend.var_units("channel_exit_water_sediment~bedload__mass_flow_rate"), '= Max Bedload Discharge')
```

```python
# write summary statistics to a csv file
import csv
# field names
fields = ['Mean Qw', 'Mean Qs', 'Mean SSC', 'Mean Qb', 'Max Qw', 'Max Qs', 'Max SSC', 'Max Qb', 'SD Qw', 'SD Qs', 'SD SSC', 'SD Qb', 'Qw 10 %ile', 'Qs 10 %ile', 'SSC 10 %ile', 'QB 10 %ile', 'Qw 20 %ile', 'Qs 20 %ile', 'SSC 20 %ile', 'QB 20 %ile','Qw 50 %ile', 'Qs 50 %ile', 'SSC 50 %ile', 'QB 50 %ile','Qw 90 %ile', 'Qs 90 %ile', 'SSC 90 %ile', 'QB 90 %ile']
# data rows of csv file
rows = [q.mean(), qs.mean(), cs.mean(), qb.mean(), q.max(), qs.max(), cs.max(), qb.max(), q.std(), qs.std(), cs.std(), qb.std(), np.percentile(q,10), np.percentile(qs,10),np.percentile(cs,10), np.percentile(qb,10), np.percentile(q,20), np.percentile(qs,20),np.percentile(cs,20), np.percentile(qb,20), np.percentile(q,50), np.percentile(qs,50),np.percentile(cs,50), np.percentile(qb,50),np.percentile(q,90), np.percentile(qs,90),np.percentile(cs,90), np.percentile(qb,90)]
# name of csv file
filename = "summ_stats_v1.csv"
# writing to csv file
with open(filename, 'w') as csvfile:
    # creating a csv writer object
    csvwriter = csv.writer(csvfile)
    # writing the fields
    csvwriter.writerow(fields)
    # writing the data rows
    csvwriter.writerow(rows)
```

## Reference Scenario Only

Now let's compare model outputs to real data (only for reference period 1975-2000)

```python
# read in CSV with recent data from Bahadurabad Bridge
import pandas as pd
Baha_csv = pd.read_csv('Baha_Qw_1976-2006_v3.csv')
print(Baha_csv)
```

```python
# create panda array from date and discharge columns
import numpy as np
import matplotlib.pyplot as plt
Date_Baha_csv = Baha_csv['Date']
Qw_Baha_csv = Baha_csv['Discharge (m3/s)']
Date_Baha_csv = [pd.to_datetime(d) for d in Date_Baha_csv]
print(Date_Baha_csv[0:5])
Date_Baha = np.array(Date_Baha_csv)
Qw_Baha = np.array(Qw_Baha_csv)
```

```python
# Create pandas dataframe
Baha_df = pd.DataFrame(Qw_Baha, Date_Baha)
Baha_df.columns = ['Discharge (m3/s)']
Baha_df.head()
```

```python
# plot recent Bahadurabad data
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
plt.plot(Date_Baha, Qw_Baha, color = 'c')
plt.title('Recent Water Discharge at Bahadurabad Bridge, Brahmaputra River (1976-2003)')
plt.xlabel('Year')
plt.xticks()
plt.ylabel('River discharge (m3/s)')
plt.show()
```

```python
# Calculate and plot *observed sediment discharge
# Brahmaputra Q-Qs relationship (1989-1994): Qs = 0.005*Q^1.56; R2 = 0.78 (Higgens et al., 2018)
Qs_Baha = (Qw_Baha**1.56)*0.005
plt.plot(Date_Baha, Qs_Baha, color = 'k')
plt.title('Recent (Rated) Sediment Discharge at Bahadurabad Bridge, Brahmaputra River (1976-2003)')
plt.xlabel('Year')
plt.xticks()
plt.ylabel('River discharge (m3/s)')
plt.show()
```

```python
# Calculate and plot mean annual flow (observed vs simulated)
import numpy as np
from numpy import reshape
# Plot observed
Qw_Reshape_Obs = Qw_Baha.reshape(27,365)
Qw_Mean_Rows_Obs = np.mean(Qw_Reshape_Obs, axis = 1)
plt.plot(Qw_Mean_Rows_Obs, color='c', label='obs')
plt.xlabel('Year in simulation')
plt.ylabel('Mean Annual Discharge (m^3/s)')
plt.title('Observed vs. Simulated Mean Annual Flow, Brahmaputra River (1976-2003)')
# Plot simulated
Qw_Reshape_Sim = q.reshape(27,365)
Qw_Mean_Rows_Sim = np.mean(Qw_Reshape_Sim, axis = 1)
plt.plot(Qw_Mean_Rows_Sim,color='tab:blue', label='sim')
plt.legend()
plt.show()
```

```python
# Plot mean daily discharge over simulation time period (observed vs simulated)
# Plot observed
Qw_Reshape_Obs = Qw_Baha.reshape(27,365)
Qw_Mean_Cols_Obs = np.mean(Qw_Reshape_Obs, axis = 0)
plt.plot(Qw_Mean_Cols_Obs, color='c', label='obs')
plt.xlabel('Day of Year')
plt.ylabel('Mean Daily Discharge (m^3/s)')
plt.title('Observed vs. Simulated Mean Daily Water Discharge, Brahmaputra River, 1976-2003')
# Plot simulated
Qw_Daily = np.mean(Qw_Reshape_Sim, axis = 0)
plt.plot(Qw_Daily, color = 'tab:blue', label='sim')
plt.legend()
plt.show()
```

```python
# Plot mean daily discharge over simulation time period (observed vs simulated)
# Plot observed
Qs_Reshape_Obs = Qs_Baha.reshape(27,365)
Qs_Mean_Rows_Obs = np.mean(Qs_Reshape_Obs, axis = 1)
plt.plot(Qs_Mean_Rows_Obs, color='k', label='obs')
plt.xlabel('Day of Year')
plt.ylabel('Mean Annual Sediment Discharge (kg/s)')
plt.title('Observed vs. Simulated Max Daily Sediment Discharge, Brahmaputra River, 1976-2003')
# Plot simulated
Qs_Reshape_Sim = qs.reshape(27,365)
Qs_Mean_Rows_Sim = np.mean(Qs_Reshape_Sim, axis = 1)
plt.plot(Qs_Mean_Rows_Sim, color = 'tab:brown', label = 'sim')
plt.legend()
plt.show()
```

```python
# Plot mean daily sediment discharge over simulation time period (observed vs simulated)
# Plot observed
Qs_Mean_Cols_Obs = np.mean(Qs_Reshape_Obs, axis = 0)
plt.plot(Qs_Mean_Cols_Obs,color='k', label='obs')
# Plot simulated
Qs_Daily = np.mean(Qs_Reshape_Sim, axis = 0)
plt.plot(Qs_Daily, color = 'tab:brown', label='sim')
plt.xlabel('Day of Year')
plt.ylabel('Mean Daily Sediment Discharge (kg/s)')
plt.title('Observed vs. Simulated Mean Daily Sediment Discharge, Brahmaputra River, 1976-2003')
plt.legend()
plt.show()
```

```python
# compare observed and simulated mean water discharge statistics
print(np.nanmean(Qw_Baha), '= Mean Observed Qw m^3 / s')
print(q.mean(),'= Mean Simulated Annual Qw', hydrotrend.var_units("channel_exit_water__volume_flow_rate"),)
print((((np.nanmean(Qw_Baha)-q.mean())/np.nanmean(Qw_Baha))*100), ' = Percent Error, Mean Qw, Simulated Vs. Observed')
print()

# compare observed and simulated mean sediment discharge statistics
print(np.nanmean(Qs_Baha), '= Mean Observed Qs m^3 / s')
print(qs.mean(),'= Mean Simulated Annual Qs', hydrotrend.var_units("channel_exit_water__volume_flow_rate"),)
print((((np.nanmean(Qs_Baha)-qs.mean())/np.nanmean(Qs_Baha))*100), ' = Percent Error, Mean Qs, Simulated Vs. Observed')
print()

# compare observed and simulated max water discharge statistics
print(np.nanmax(Qw_Mean_Cols_Obs), '= Max Annual Observed Daily Qw')
print(Qw_Daily.max(),'= Max Annual Simulated Daily Qw')
print(((np.nanmax(Qw_Mean_Cols_Obs)-Qw_Daily.max())/np.nanmax(Qw_Mean_Cols_Obs)*100), '= Percent Error, Max Qw, Simulated Vs. Observed')
Qw_Max_Index_Col_Obs = np.nanargmax(Qw_Mean_Cols_Obs, axis=0)
print(Qw_Max_Index_Col_Obs, '= Day of Year of Observed Peak Flow')
Qw_Max_Index_Col_Sim = np.nanargmax(Qw_Daily, axis=0)
print(Qw_Max_Index_Col_Sim, '= Day of Year of Simulated Peak Flow')
print()

# compare observed and simulated max water discharge statistics
print(np.nanmax(Qs_Mean_Cols_Obs), '= Max Annual Observed Daily Qs')
print(Qs_Daily.max(),'= Max Annual Simulated Daily Qs')
print(((np.nanmax(Qs_Mean_Cols_Obs)-Qs_Daily.max())/np.nanmax(Qs_Mean_Cols_Obs)*100), '= Percent Error, Max Qs, Simulated Vs. Observed')
Qs_Max_Index_Col_Obs = np.nanargmax(Qs_Mean_Cols_Obs, axis=0)
print(Qs_Max_Index_Col_Obs, '= Day of Year of Observed Peak Qs')
Qs_Max_Index_Col_Sim = np.nanargmax(Qs_Daily, axis=0)
print(Qs_Max_Index_Col_Sim, '= Day of Year of Simulated Peak Qs')
print()

# additional statistical measurements
print(np.corrcoef(Qw_Baha, q), '= Correlation Coefficient, Qw, Simulated Vs. Observed')
print(np.cov(Qw_Baha, q), '= Covarience, Qw, Simulated Vs. Observed')
print(np.corrcoef(Qs_Baha, q), '= Correlation Coefficient, Qs, Simulated Vs. Observed')
print(np.cov(Qs_Baha, q), '= Covarience, Qs, Simulated Vs. Observed')
print(np.nanstd(Qw_Baha), '= Std. Dev, Observed Qw')
print(np.nanstd(Qs_Baha), '= Std. Dev, Observed Qs')
```
