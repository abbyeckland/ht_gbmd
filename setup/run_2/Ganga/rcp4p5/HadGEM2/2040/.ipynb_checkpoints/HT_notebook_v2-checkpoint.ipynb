```python
# Import numpy and matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Import the pymt package 
import pymt.models
# Create new instance. With each new run, it is wise to rename the instance.
hydrotrend = pymt.models.Hydrotrend()
```

```python
# Add directory for output files
config_file, config_folder = "hydro_config.txt", "Hydrotrend"
```

```python
cat  Hydrotrend/HYDRO.IN
```

```python
cat  Hydrotrend/HYDRO0.HYPS
```

```python
cat  Hydrotrend/HYDRO.CLIMATE
```

```python
#In pymt one can always find out what output a model generates by using the .output_var_names method. 
hydrotrend.output_var_names
```

```python
# Now we initialize the model with the configure file and in the configure folder
hydrotrend.initialize(config_file, config_folder)
```

```python
# this line of code lists time parameters: when, how long, and at what timestep the model simulation will work
hydrotrend.start_time, hydrotrend.time, hydrotrend.end_time, hydrotrend.time_step, hydrotrend.time_units
```

```python
# this code declares numpy arrays for several important parameters we want to save.
# empty is declaring space in memory for info to go in
n_days = int(hydrotrend.end_time)
q = np.empty(n_days) # river discharge at the outlet
qs = np.empty(n_days)# sediment load at the outlet
cs = np.empty(n_days) # suspended sediment concentration for different grainsize classes at the outlet
qb = np.empty(n_days) # bedload at the outlet
# here we have coded up the time loop using i as the index
# we update the model with one timestep at the time, untill we reach the end time 
# for each time step we also get the values for the output parameters we wish to 
for i in range(n_days):
    hydrotrend.update()
    q[i] = hydrotrend.get_value("channel_exit_water__volume_flow_rate")
    qs[i] = hydrotrend.get_value("channel_exit_water_sediment~suspended__mass_flow_rate")
    cs[i] = hydrotrend.get_value("channel_exit_water_sediment~suspended__mass_concentration")
    qb[i] = hydrotrend.get_value("channel_exit_water_sediment~bedload__mass_flow_rate")
```

```python
# Plot river discharge (q)
plt.plot(q, color='tab:blue')
plt.title('Simulated Qw')
plt.xlabel('Day in simulation')
plt.ylabel('River discharge (m3/s)')
plt.show()
```

```python
import numpy as np
from numpy import reshape
# Plot mean annual discharge
Qw_Reshape = q.reshape(30,365)
Qw_Mean_Rows = np.mean(Qw_Reshape, axis = 1)
plt.plot(Qw_Mean_Rows, color='c', label='obs')
plt.xlabel('Year in simulation')
plt.ylabel('Mean Annual Discharge (m^3/s)')
plt.title('Simulated Mean Annual Flow')
plt.show()
```

```python
# Plot mean daily discharge 
Qw_Reshape = q.reshape(30,365)
Qw_Mean_Cols = np.mean(Qw_Reshape, axis = 0)
plt.plot(Qw_Mean_Cols, color='c', label='obs')
plt.xlabel('Day of Year')
plt.ylabel('Mean Daily Discharge (m^3/s)')
plt.title('Simulated Mean Daily Water Discharge')
plt.show()

Qw_Max_Index_Col = np.nanargmax(Qw_Mean_Cols, axis=0)
print(Qw_Max_Index_Col, '= Day of Year of Peak Qs')
```

```python
# Plot suspended sediment discharge (qs)
plt.plot(qs,color='tab:brown')
plt.title('Simulated Qs')
plt.xlabel('Day in simulation')
plt.ylabel('Sediment load (kg/s)')
plt.show()
```

```python
# Plot mean daily sediment discharge
Qs_Reshape = qs.reshape(30,365)
Qs_Mean_Rows = np.mean(Qs_Reshape, axis = 1)
plt.plot(Qs_Mean_Rows, color='tab:brown', label='obs')
plt.xlabel('Day of Year')
plt.ylabel('Mean Annual Sediment Discharge (kg/s)')
plt.title('Simulated Max Daily Sediment Discharge')
plt.show()
```

```python
Qs_Mean_Cols = np.mean(Qs_Reshape, axis = 0)
plt.plot(Qs_Mean_Cols, color = 'tab:brown', label='sim')
plt.xlabel('Day of Year')
plt.ylabel('Mean Daily Sediment Discharge (kg/s)')
plt.title('Simulated Mean Daily Sediment Discharge')
plt.show()

Qs_Max_Index_Col = np.nanargmax(Qs_Mean_Cols, axis=0)
print(Qs_Max_Index_Col, '= Day of Year of Peak Qs')
```

```python
# Print summary statistics from model run
print('Mean')
print(q.mean(), hydrotrend.var_units("channel_exit_water__volume_flow_rate"), '= Mean Water Discharge')
print(qs.mean(), hydrotrend.var_units("channel_exit_water_sediment~suspended__mass_flow_rate"), '= Mean Suspended Sediment Discharge')
print(cs.mean(), hydrotrend.var_units("channel_exit_water_sediment~suspended__mass_concentration"), '= Mean Suspended Sediment Concentration')
print(qb.mean(), hydrotrend.var_units("channel_exit_water_sediment~bedload__mass_flow_rate"), '= Mean Bedload Discharge')
print()
print('Max')
print(q.max(), hydrotrend.var_units("channel_exit_water__volume_flow_rate"), '= Max Water Discharge')
print(qs.max(), hydrotrend.var_units("channel_exit_water_sediment~suspended__mass_flow_rate"), '= Max Suspended Sediment Discharge')
print(cs.max(), hydrotrend.var_units("channel_exit_water_sediment~suspended__mass_concentration"), '= Max Suspended Sediment Concentration')
print(qb.max(), hydrotrend.var_units("channel_exit_water_sediment~bedload__mass_flow_rate"), '= Max Bedload Discharge')
print()
```

```python
# write summary statistics to a csv file
import csv
# field names
fields = ['Mean Qw', 'Mean Qs', 'Mean SSC', 'Mean Qb', 'Max Qw', 'Max Qs', 'Max SSC', 'Max Qb', 'SD Qw', 'SD Qs', 'SD SSC', 'SD Qb', 'Qw 10 %ile', 'Qs 10 %ile', 'SSC 10 %ile', 'QB 10 %ile', 'Qw 20 %ile', 'Qs 20 %ile', 'SSC 20 %ile', 'QB 20 %ile','Qw 50 %ile', 'Qs 50 %ile', 'SSC 50 %ile', 'QB 50 %ile','Qw 90 %ile', 'Qs 90 %ile', 'SSC 90 %ile', 'QB 90 %ile', 'DOY Peak Qw', 'DOY Peak Qs']
# data rows of csv file
rows = [q.mean(), qs.mean(), cs.mean(), qb.mean(), q.max(), qs.max(), cs.max(), qb.max(), q.std(), qs.std(), cs.std(), qb.std(), np.percentile(q,10), np.percentile(qs,10),np.percentile(cs,10), np.percentile(qb,10), np.percentile(q,20), np.percentile(qs,20),np.percentile(cs,20), np.percentile(qb,20), np.percentile(q,50), np.percentile(qs,50),np.percentile(cs,50), np.percentile(qb,50),np.percentile(q,90), np.percentile(qs,90), np.percentile(cs,90), np.percentile(qb,90), Qw_Max_Index_Col, Qs_Max_Index_Col]
# name of csv file
filename = "summ_stats_v2.csv"
# writing to csv file
with open(filename, 'w') as csvfile:
    # creating a csv writer object
    csvwriter = csv.writer(csvfile)
    # writing the fields
    csvwriter.writerow(fields)
    # writing the data rows
    csvwriter.writerow(rows)
```

## Reference Scenario Only

Now let's compare model outputs to real data (only for reference period 1975-2000)

```python
# read in CSV with recent data from Hardinge Bridge
import pandas as pd
Hardinge = pd.read_csv("Ganges/Q0/V1/Q0_1975_2000.csv")
print(Hardinge)
```

```python
# create panda array from date and discharge columns
import matplotlib.pyplot as plt
Date_Hardinge = Hardinge['Date']
Q_Hardinge = Hardinge['Discharge (m3/s)']
Date_Hardinge = [pd.to_datetime(d) for d in Date_Hardinge]
print(Date_Hardinge[0:5])
date = np.array(Date_Hardinge)
q_hardinge = np.array(Q_Hardinge)
```

```python
Hardinge_GQ0V1_df = pd.DataFrame(q_hardinge, date)
Hardinge_GQ0V1_df.columns = ['Discharge (m3/s)']
Hardinge_GQ0V1_df.head()
```

```python
# plot recent Hardinge data
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
plt.plot(date, q_hardinge, color = 'c')
plt.title('Recent Water Discharge at Hardinge Bridge, Ganges River (1975-2000)')
plt.xlabel('Year')
plt.xticks()
plt.ylabel('River discharge (m3/s)')
plt.show()
```

```python
# Calculate and plot *observed sediment discharge
# *Ganges Q-Qs relationship (1980-1995): Qs = 0.007*Q^1.51; R2 = 0.68 (Higgens et al., 2018)
qs_hardinge = (q_hardinge**1.51)*0.007
plt.plot(date, qs_hardinge, color = 'k')
plt.title('Recent (Rated) Sediment Discharge at Hardinge Bridge, Ganges River (1975-2000)')
plt.xlabel('Year')
plt.xticks()
plt.ylabel('River discharge (m3/s)')
plt.show()
```

```python
# Calculate and plot mean annual flow (observed vs simulated)
import numpy as np
from numpy import reshape
# Plot observed
q_reshape_obs = q_hardinge.reshape(25,365)
q_mean_rows_obs = np.mean(q_reshape_obs, axis = 1)
plt.plot(q_mean_rows_obs,color='c', label='obs')
plt.xlabel('Year in simulation')
plt.ylabel('Mean Annual Discharge (m^3/s)')
plt.title('Observed vs. Simulated Mean Annual Flow, Ganges River (1975-2000)')
# Plot simulated
q_reshape = q.reshape(25,365)
q_mean_rows = np.mean(q_reshape, axis = 1)
plt.plot(q_mean_rows,color='tab:blue', label='sim')
plt.legend()
plt.show()

#print observed water discharge statistics and compare with simulations
print(Q_Hardinge.mean(), '= Mean Observed River Discharge m^3 / s')
print(q.mean(),'= Mean Simulated Annual Water Discharge', hydrotrend.get_var_units("channel_exit_water__volume_flow_rate"),)
print((((Q_Hardinge.mean()-q.mean())/Q_Hardinge.mean())*100), ' = Percent Error, Mean Q, Simulated Vs. Observed')
```

```python
# Calculate and plot mean annual flow (observed vs simulated)
import numpy as np
from numpy import reshape
# Plot observed
q_reshape_obs = q_hardinge.reshape(25,365)
q_max_rows_obs = np.max(q_reshape_obs, axis = 1)
plt.plot(q_max_rows_obs,color='c', label='obs')
plt.xlabel('Year in simulation')
plt.ylabel('Max Annual Discharge (m^3/s)')
plt.title('Observed vs. Simulated Max Annual Flow, Ganges River (1975-2000)')
# Plot simulated
q_reshape = q.reshape(25,365)
q_max_rows = np.max(q_reshape, axis = 1)
plt.plot(q_max_rows,color='tab:blue', label='sim')
plt.legend()
plt.show()

print(Q_Hardinge.max(), '= Max Observed River Discharge m^3 / s')
print(q.max(), '= Max Simulated Annual Water Discharge', hydrotrend.get_var_units("channel_exit_water__volume_flow_rate"),)
print((((Q_Hardinge.max()-q.max())/Q_Hardinge.max())*100), ' = Percent Error, Max Q, Simulated Vs. Observed')
```

```python
# Plot mean daily discharge over 25 year period (observed vs simulated)
# Plot observed
q_mean_cols_obs = np.mean(q_reshape_obs, axis = 0)
plt.plot(q_mean_cols_obs,color='c', label='obs')
# Plot simulated
q_daily = np.mean(q_reshape, axis = 0)
plt.plot(q_daily, color = 'tab:blue', label='sim')
plt.xlabel('Day of Year')
plt.ylabel('Mean Daily Discharge (m^3/s)')
plt.title('Observed vs. Simulated Mean Daily Water Discharge, Ganges River, 1975-2000')
plt.legend()
plt.show()

print(q_mean_cols_obs.max(), '= Max annual observed daily discharge')
print(q_daily.max(),'= Max annual simulated daily discharge')
print(((q_mean_cols_obs.max()-q_daily.max())/q_mean_cols_obs.max()*100), '= Percent Error')

q_max_index_col_obs = np.argmax(q_mean_cols_obs, axis=0)
print(q_max_index_col_obs, '= Day of Year of Observed Peak Flow')
q_max_index_col_sim = np.argmax(q_daily, axis=0)
print(q_max_index_col_sim, '= Day of Year of Simulated Peak Flow')
```

```python
# Plot mean daily discharge over 25 year period (observed vs simulated)
# Plot observed
q_max_cols_obs = np.max(q_reshape_obs, axis = 0)
plt.plot(q_max_cols_obs,color='c', label='obs')
# Plot simulated
q_daily = np.max(q_reshape, axis = 0)
plt.plot(q_daily, color = 'tab:blue', label='sim')
plt.xlabel('Day of Year')
plt.ylabel('Max Daily Discharge (m^3/s)')
plt.title('Observed vs. Simulated Max Daily Water Discharge, Ganges River, 1975-2000')
plt.legend()
plt.show()
```

```python
# Calculate and plot mean annual sediment discharge (observed vs. simulated)
# Plot observed
qs_reshape_obs = qs_hardinge.reshape(25,365)
qs_mean_rows_obs = np.mean(qs_reshape_obs, axis = 1)
plt.plot(qs_mean_rows_obs,color='k', label='obs')
plt.xlabel('Year in simulation')
plt.ylabel('Mean Annual Sediment Discharge (kg/s)')
plt.title('Observed vs. Simulated Mean Annual Sediment Discharge, Ganges River (1975-2000)')
# Plot simulated
qs_reshape = qs.reshape(25,365)
qs_mean_rows = np.mean(qs_reshape, axis = 1)
plt.plot(qs_mean_rows,color='tab:brown', label='sim')
plt.legend()
plt.show()
```

```python
# Plot mean daily sediment discharge over 25 year period (observed vs simulated)
# Plot observed
qs_mean_cols_obs = np.mean(qs_reshape_obs, axis = 0)
plt.plot(qs_mean_cols_obs,color='k', label='obs')
# Plot simulated
qs_daily = np.mean(qs_reshape, axis = 0)
plt.plot(qs_daily, color = 'tab:brown', label='sim')
plt.xlabel('Day of Year')
plt.ylabel('Mean Daily Sediment Discharge (kg/s)')
plt.title('Observed vs. Simulated Mean Daily Sediment Discharge, Ganges River, 1975-2000')
plt.legend()
plt.show()

qs_max_index_col_obs = np.argmax(qs_mean_cols_obs, axis=0)
print(qs_max_index_col_obs, '= Day of Year of Observed Peak Qs')
qs_max_index_col_sim = np.argmax(qs_daily, axis=0)
print(qs_max_index_col_sim, '=Day of Year of Simulated Peak Qs')
```

```python
#print observed water discharge statistics and compare with simulations
print(qs_hardinge.mean(), '= Mean Observed Sediment Discharge kg / s')
print(qs.mean(),'= Mean Simulated Annual Sediment Discharge', hydrotrend.get_var_units("channel_exit_water_sediment~suspended__mass_flow_rate"),)
print((((qs_hardinge.mean()-qs.mean())/qs_hardinge.mean())*100), ' = Percent Error, Mean Qs, Simulated Vs. Observed')
print(qs_hardinge.max(), '= Max Observed Sediment Discharge kg / s')
print(qs.max(), '= Max Simulated Annual Sediment Discharge', hydrotrend.get_var_units("channel_exit_water_sediment~suspended__mass_flow_rate"),)
print((((qs_hardinge.max()-qs.max())/qs_hardinge.max())*100), ' = Percent Error, Max Qs, Simulated Vs. Observed')
```
