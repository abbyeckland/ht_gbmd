```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from numpy import reshape
import pylab
```

## Import Obs vs. Sim Reference Data

```python
gfdl = pd.DataFrame(pd.read_csv('/Users/aeckland/hydrotrend_gbmd/setup/run_1/Ganga/reference/GFDL-ESM2M/REF_summ_stats_v3.csv'))
hadgem = pd.DataFrame(pd.read_csv('/Users/aeckland/hydrotrend_gbmd/setup/run_1/Ganga/reference/HadGEM2/REF_summ_stats_v3.csv'))
ipsl = pd.DataFrame(pd.read_csv('/Users/aeckland/hydrotrend_gbmd/setup/run_1/Ganga/reference/IPSL_CM5A/REF_summ_stats_v3.csv'))
miroc = pd.DataFrame(pd.read_csv('/Users/aeckland/hydrotrend_gbmd/setup/run_1/Ganga/reference/MIROC-ESM-CHEM/REF_summ_stats_v3.csv'))
noresm = pd.DataFrame(pd.read_csv('/Users/aeckland/hydrotrend_gbmd/setup/run_1/Ganga/reference/NORESM-M/REF_summ_stats_v3.csv'))
```

## Plot Obs vs. Sim

```python
plt.figure(figsize = (14,10))
models = ['GFDL-ESM2M', 'HadGEM2', 'IPSL_CM5A', 'MIROC-ESM-CHEM', 'NORESM-M']
obs_qw_mean = concat['Mean Obs Qw']
obs_qw_max = concat['Max Obs Qw']
plt.bar(models, concat['Max Sim Qw'], width = 0.2, color = 'deepskyblue', label = 'Max Sim Qw')
plt.bar(models, concat['Mean Sim Qw'], width = 0.2, color = 'paleturquoise',  label = 'Mean Sim Qw')
plt.plot(models, obs_qw_max, color = 'deepskyblue', linewidth = 3, linestyle = 'dashed', label = 'Max Obs Qw')
plt.plot(models, obs_qw_mean, color = 'paleturquoise', linewidth = 3, linestyle = 'dashed', label = 'Mean Obs Qw')
plt.legend()
plt.title('Observed vs. Simulated Mean & Max Qw, 1976-2002, Ganges River', size = 18)
plt.xlabel('Climate Model', size = 14)
plt.ylabel('Water Discharge (m$^3/$s)', size = 14)

plt.savefig('/Users/aeckland/hydrotrend_gbmd/figures/v1/validation/ganga_qw_mean_max_bars.png')

plt.show()
```

## Import Discharge: Observed

```python
obs_qw = pd.DataFrame(pd.read_csv('/Users/aeckland/hydrotrend_gbmd/water_data/Ganges/Hardinge_Qw_1976-2006.csv'))
obs_qw = np.array(obs_qw['Discharge (m3/s)'])
```

```python
# Reshape array (mean daily)

obs_qw_reshape_qw = obs_qw.reshape(30, 365)
obs_qw_mean_cols = np.mean(obs_qw_reshape_qw, axis = 0)

# Reshape array (mean annual)

obs_qw_reshape_qw = obs_qw.reshape(30, 365)
obs_qw_mean_rows = np.mean(obs_qw_reshape_qw, axis = 1)
```
